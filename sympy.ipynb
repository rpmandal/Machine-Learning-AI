```python
x = sp.symbols('x')
expr = sp.sin(x) * sp.exp(x)
sp.plot(expr, (x, -3, 3))
```

```python
sp.init_printing()
x,y,z,theta = sp.symbols('x,y,z,\\theta')
expr  = sp.sin(theta)*sp.exp(y)*sp.log(z)*(x+y*theta)**4
expr = sp.Integral(expr,(x,0,2))
expr = sp.Integral(expr,(y,-1,1))
expr = sp.Integral(expr,(z,-2,0))
expr = sp.Derivative(expr,theta)
expr
```
