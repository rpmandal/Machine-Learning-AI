```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv('C:/Python/Missing_Values/EDA_census.csv')
print(df.info())

```

```python
# =============================================================================
# What percentage of females in the age group 20-24 are illiterate in India, 
# i.e. out of all the females in the age group 20-24, what fraction is illiterate?
# =============================================================================

#Let's first check what the is total no of females in age group 20-24 in INDIA
Age_20_24 = df[(df['Age']=='20-24') & (df['Area']=='INDIA') & (df['T_R_U']=='Total')]
Total_F_20_24 = Age_20_24.Total_Females #53839529
Total_Illeterate_F_20_24 = Age_20_24.Total_Illiterate_Females #12212466
Percentage_Illeterate_F_20_24 = (Total_Illeterate_F_20_24/Total_F_20_24)*100
Total_Illeterate_F_20_24
```

```python
# =============================================================================
# Compare the literacy rates (i.e. the number of literates / total number of  
# population)in each age group and choose the correct option.
# =============================================================================
All_Age_Subset = df[(df['Area']=='INDIA') & (df['T_R_U']=='Total') & ~(df['Age']=='All ages')]
All_Age_Subset['Literacy_Percentage'] = (All_Age_Subset['Total_Literate_Persons']/All_Age_Subset['Total_Persons'])*100
#All_Age_Subset
```

```python
# Which state shows the highest female literacy rate?
# =============================================================================
State_Wise_Total_Subset = df[(df['T_R_U']=='Total') & (df['Age']=='All ages') & ~(df['Area']=='INDIA')]
State_Wise_Total_Subset['Female_Literacy_Percentage'] = (State_Wise_Total_Subset['Total_Literate_Females']/State_Wise_Total_Subset['Total_Females']) * 100
#State_Wise_Total_Subset
```

```python
#setting seaborn theme
sns.set(style = "white")
# Set the figure size
plt.figure(figsize=(70,12))
plt.suptitle('Female Literacy percentage accross states', fontsize=14, fontweight='bold')

sns.barplot(x='Area',y='Female_Literacy_Percentage', data=State_Wise_Total_Subset)
#plt.xlabel('Country')
#plt.ylabel('Total amount invested')
```
