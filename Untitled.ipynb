```python
import numpy as np
import pandas as pd

data_df = pd.DataFrame({'name':['Kari0','Ku0l','0kul', 'Saif0.0','Kajal'], 
                        'age':[22,25,24,28,32], 
                        'occupation':['engineer','doctor','data analyst','teacher','teacher']})
def correctSpell(x) :
    if x[:1] == '0':
        x =  'Na' + x[1:]
        return x
    elif x[-1] == '0':
        return x
    else:
        x = x.replace('0','na')
        return x

data_df['name'] = data_df['name'].apply(correctSpell)
data_df
```

```python
mapping = pd.read_csv("/Investment Analysis/mapping.csv")
mapping_df = pd.DataFrame(mapping)
mapping_df['category_list'] = mapping_df['category_list'].apply(correctSpell)
```
