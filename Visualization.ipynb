```python
# Axis labels and title
plt.plot(x, y)

# x and y labels, and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

# Define the range of labels of the axis 
# Arguments: plt.axis(xmin, xmax, ymin, ymax)
plt.xlim([20, 80])
plt.ylim([20, 800])
plt.show()
```

```python
# Change the colors and line type

# initialising x and y arrays
x = np.linspace(0, 10, 20)
y = x*2
#y = x = np.linspace(0, 10, 20)

# color blue, line type '+'
plt.plot(x, y, 'r*')

# put x and y labels, and the title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

```python
x = np.linspace(1,10,10)
#y1 = np.log(x)
#y2 = np.log(x)**2
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.sqrt(x)
plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y1,'ks')
plt.axis("equal")
plt.title("$y_1 = sin(x)$")
plt.grid()
#plt.show()

plt.subplot(2,2,2)
plt.plot(x,y2,'b--')
plt.axis("equal")
#plt.title("$y_2 = log(x)^2$")
plt.title("$y_2 = cos(x)$")
plt.grid()
plt.show()

plt.subplot(2,2,3)
plt.plot(x,y3,'r')
plt.axis("equal")
#plt.title("$y_2 = log(x)^2$")
plt.title("$y_3 = x^2$")
plt.grid()
#plt.show()

plt.subplot(2,2,4)
plt.plot(x,y3,'g')
plt.axis("equal")
#plt.title("$y_2 = log(x)^2$")
plt.title("$y_4 = \sqrt{x}$")
plt.grid()
plt.show()


```
